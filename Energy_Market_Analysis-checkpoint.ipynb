```python
# 3.1 Data Loading Code

import pandas as pd
import yfinance as yf
from pathlib import Path

# ------------------------------------------------------------------
# Configuration
# ------------------------------------------------------------------

DATA_DIR = Path("Data")
start_date = "2015-01-01"
end_date = "2025-01-01"

# ------------------------------------------------------------------
# Spot Prices (CSV Sources - Weekly)
# ------------------------------------------------------------------

brent_spot = pd.read_csv( DATA_DIR / "BrentSpotPrices.csv", parse_dates=["Date"], index_col="Date")[["Price"]]
WTI_spot = pd.read_csv( DATA_DIR / "WTISpotPrices.csv", parse_dates=["Date"], index_col="Date")[["Price"]]

# ------------------------------------------------------------------
# Spot Prices (Daily to Weekly Conversion)
# ------------------------------------------------------------------

henry_hub_spot = pd.read_csv( DATA_DIR / "HenryHubSpotPrice.csv", parse_dates=["observation_date"],index_col="observation_date")
henry_hub_weekly = henry_hub_spot.resample("W-FRI").last()

diesel_spot = pd.read_csv( DATA_DIR / "DieselFuelPricesSpot.csv", parse_dates=["observation_date"],index_col="observation_date")
diesel_weekly = diesel_spot.resample("W-FRI").last()

jet_fuel_spot = pd.read_csv( DATA_DIR / "JetFuelSpotPrice.csv", parse_dates=["observation_date"],index_col="observation_date")
jet_fuel_weekly = jet_fuel_spot.resample("W-FRI").last()

# ------------------------------------------------------------------
# Futures Prices (Yahoo Finance - Weekly)
# ------------------------------------------------------------------

brent_future = yf.download("BZ=F" , start=start_date, end=end_date, interval="1wk")[["Close"]].rename(columns={"Close":"Brent_F"})
henry_hub_future = yf.download("NG=F", start=start_date, end=end_date, interval="1wk")[["Close"]].rename(columns={"Close": "henry_hub_future"})
TTF_future = yf.download("TTF=F", start=start_date, end=end_date, interval="1wk")[["Close"]].rename(columns={"Close": "TTF_F"})
gasoline_future = yf.download("RB=F", start=start_date, end=end_date, interval="1wk")[["Close"]].rename(columns={"Close": "Gasoline_F"})

# ------------------------------------------------------------------
# Future Prices (CSV Sources)
# ------------------------------------------------------------------

dubai_oil_future = pd.read_csv( DATA_DIR / "DubaiCrudeOilPlattsFuturePrices.csv", parse_dates=["Date"], index_col="Date")[["Price"]]
JKM_future = pd.read_csv( DATA_DIR / "JKMFuturePrice.csv", parse_dates=["Date"], index_col="Date")[["Price"]]

# ------------------------------------------------------------------
# Data Validation
# ------------------------------------------------------------------

datasets = { "Brent Spot": brent_spot, "WTI Spot": WTI_spot, "Henry Hub Spot (Weekly)": henry_hub_weekly, "Diesel Spot (Weekly)": diesel_weekly,
            "Jet Fuel Spot (Weekly)": jet_fuel_weekly, "Brent Futures": brent_future, "Henry Hub Futures": henry_hub_future, 
            "TTF Futures": TTF_future, "Gasoline Futures": gasoline_future, "Dubai Futures": dubai_oil_future, "JKM Futures": JKM_future}

print("Data Validation Summary:\n")

for name, df in datasets.items():
    print(f"{name:<30} | Shape: {df.shape}")
```

# 4.0 Data Cleaning and Preparation 

In this section, spot and futures price series are consolidated into two aligned datasets to enable consistent preprocessing and analysis. All series are standardised with respect to column naming, data types, chronological ordering, and weekly frequency.

Duplicate entries, formatting inconsistencies, and obvious anomalies are addressed as part of the cleaning process. Preliminary diagnostics, including summary statistics and visual inspection, are used to verify internal consistency and identify data gaps.

Heatmaps of missing values are employed to provide a clear overview of historical data availability across commodities and markets. This step establishes the structure of missingness prior to formal imputation and informs subsequent methodological choices.

```python
# 4.1 Data Cleaning and Preparation Code

import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# Spot Prices Table
# ------------------------------------------------------------------

# Standardise Column names
brent_spot = brent_spot.rename(columns={"Price": "Brent_S"})
WTI_spot = WTI_spot.rename(columns={"Price" : "WTI_S"})
henry_hub_weekly = henry_hub_weekly.rename(columns={"DHHNGSP" : "HenryH_S"})
diesel_weekly = diesel_weekly.rename(columns={"DDFUELUSGULF": "Diesel_S"})
jet_fuel_weekly = jet_fuel_weekly.rename(columns={"WJFUELUSGULF": "JetF_S"})

# Combine all spot prices into a single DataFrame
spots_table = pd.concat([brent_spot[["Brent_S"]], WTI_spot[["WTI_S"]], henry_hub_weekly[["HenryH_S"]], diesel_weekly[["Diesel_S"]], jet_fuel_weekly[["JetF_S"]]], axis=1)
spots_table.sort_index(inplace=True)

# ------------------------------------------------------------------
# Futures Prices Table
# ------------------------------------------------------------------

# Standardise column names
brent_future = brent_future.rename(columns={"brent_future": "Brent_F"})
henry_hub_future = henry_hub_future.rename(columns={"henry_hub_future": "HenryH_F"})
TTF_future = TTF_future.rename(columns={"TTF_future": "TTF_F"})
gasoline_future = gasoline_future.rename(columns={"gasoline_future": "Gasoline_F"})
dubai_oil_future = dubai_oil_future.rename(columns={"Price" : "Dubai_F"})
JKM_future = JKM_future.rename(columns={"Price" : "JKM_F"})

# Combine all futures prices into a single DataFrame
futures_table = pd.concat([brent_future[["Brent_F"]], henry_hub_future[["HenryH_F"]], TTF_future[["TTF_F"]], gasoline_future[["Gasoline_F"]], dubai_oil_future[["Dubai_F"]], JKM_future[["JKM_F"]]], axis=1)
futures_table.sort_index(inplace=True)
futures_table.columns = [col[0] if isinstance(col, tuple) else col for col in futures_table.columns]

futures_table = futures_table.rename(columns={"D": "Dubai_F"})
futures_table = futures_table.rename(columns={"J": "JKM_F"})

# ------------------------------------------------------------------
# Visualise Missing Values
# ------------------------------------------------------------------

# Spot Prices Heatmap 
plt.figure(figsize=(12,6))
plt.imshow(spots_table.isnull(), aspect="auto", cmap="Blues_r", interpolation="none")
plt.title("Missing Values Heatmap - Spot Prices (White = Present , Blue = Missing)")
plt.ylabel("Weeks")
plt.xticks( ticks=range(len(spots_table.columns)), labels=spots_table.columns, rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Future Prices Heatmap
plt.figure(figsize=(12,6))
plt.imshow(futures_table.isnull(), aspect="auto", cmap="Blues_r", interpolation="none")
plt.title("Missing Values Heatmap - Future Prices (White = Present , Blue = Missing)")
plt.ylabel("Weeks")
plt.xticks( ticks=range(len(futures_table.columns)), labels=futures_table.columns, rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

# 5.0 Handling missing data 

The weekly commodity price dataset contains limited gaps arising from differences in trading calendars, data publication schedules, and contract conventions across markets. These gaps were quantified for both spot and futures series to assess data coverage prior to imputation.

**Imputation Strategy**
Missing observations were primarily addressed using time-based linear interpolation, which estimates values based on surrounding observations while preserving underlying price trends and short-term market dynamics. This approach is well suited to weekly energy price data, where price evolution is typically smooth between adjacent observations.

For isolated edge cases at the beginning or end of a series—where interpolation is not feasible—a limited forward or backward fill was applied as a fallback.
For one selected futures series, an ARIMA-based imputation was implemented as a model-driven alternative. The resulting estimates were broadly consistent with the interpolated values, providing additional confidence in the robustness of the chosen imputation approach.

**Validation**
Imputed values were visually compared against the original price series to confirm that filled observations remained realistic and aligned with observed market behaviour. This validation step ensured that the imputation process did not introduce artificial trends or distort volatility patterns.

Overall, the imputation procedure produced a clean and internally consistent dataset, providing a reliable foundation for the exploratory analysis, regression modelling, and spread analysis presented in subsequent sections. 


```python
# 5.1 Handling Missing Data Code 

import pandas as pd

# ------------------------------------------------------------------
# Count Missing Values
# ------------------------------------------------------------------

# Spot prices
missing_count_spots = spots_table.isnull().sum()
print("Missing values per spot:\n")
print(missing_count_spots)

# Future prices
missing_count_futures = futures_table.isnull().sum()
print("Missing values per future:\n")
print(missing_count_futures)

# ------------------------------------------------------------------
# Percentage of Missing Values
# ------------------------------------------------------------------

# Spot percentage 
missing_percent_spots = (spots_table.isnull().sum() / len(spots_table)) * 100
print("\nPercentage of missing spot values:\n")
print(missing_percent_spots.round(2))

# Future percentage
missing_percent_futures = (futures_table.isnull().sum() / len(futures_table)) * 100
print("\nPercentage of missing future values:\n")
print(missing_percent_futures.round(2))
```

# 5.3 Filling Missing Data

### **Spot Prices** 
The spot price sries exhibited a relative high proportion of missing values (approximately 50-67%), but the gaps were sufficiently distributed to remain suitable for time-based linear interpolation. Given the weekly frequency and the generally smooth evolution of spot prices, interpolation was deemed appropiate for reconstruction missing observation while preserving underlying trends.

All missing values in the spot prices table were filled using linear interpolation, resulting in a complete dataset ready for subsequent analysis.

### **Future Prices** 
Following the assessment of missing data percentages, a differentiated strategy was adopted for the futures price series:

1. **Exclusion of Highly Sparse Series**
   JKM_F, TTF_F, and WTI_F were excluded from further analysis due to extreme data sparsity, rendering        them unsuitable for meaningful imputation or modelling.
2. **Predictive Imputation Using ARIMA**
   The HenryH_F series, despite a high proportion of missing values, was explored using an ARIMA model to
   demonstrate a predictive imputation approach and assess its consistency with similar methods.
3. **Interpolation-Based Imputation**
   For Brent_F, Gasoline_F, and Dubai_F, linear interpolation was applied to fill internal gaps, with         limited forward and backward filling used only at the series boundaries where interpolation was not        feasible. This approach preserves medium-term trends while maintaining realistic price dynamics.

The resulting completed futures dataset was used for all subsequent analysis.

Overall, this approach ensures transparent, data-driven handling of missing futures prices, while demonstrating both interpolation-based and model-based imputation techniques.

```python
# 5.4 Filling Missing Data Code

# Imputation and Visual Validation
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# ------------------------------------------------------------------
# Linear Interpolation for Spot Prices
# ------------------------------------------------------------------

# Apply time-based interpolation and edge filling
spots_interpolated = spots_table.interpolate(method="time")
spots_filled = spots_interpolated.ffill().bfill()

# Visual validation for a representative spot 
commodity = "Brent_S"

plt.figure(figsize=(14,6))
plt.plot(spots_table[commodity], label="Original", color="orange")
plt.plot(spots_filled[commodity], label="Interpolated + Fill", color="blue", alpha = 0.7)
plt.scatter( spots_table.index[spots_table[commodity].isnull()],
            spots_filled[commodity][spots_table[commodity].isnull()],
            color="red", label="Imputed Values", zorder=5)
plt.title(f"{commodity} Weekly Spot Prices - Missing Values Handled")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()


# ------------------------------------------------------------------
# Linear Interpolation for Selected Futures Prices
# ------------------------------------------------------------------

futures_to_fill = futures_table[["Brent_F", "Gasoline_F", "Dubai_F"]]

# Apply interpolation and edge filling
futures_interpolated = futures_to_fill.interpolate(method="time")
futures_filled = futures_interpolated.ffill().bfill()

# Visual Validation for a representative future
commodity = "Brent_F"

plt.figure(figsize=(14,6))
plt.plot(futures_table[commodity], label="Original", color="orange")
plt.plot(futures_filled[commodity], label="Interpolated + Edge Fill", color="blue", alpha=0.7)
plt.scatter(
    futures_table.index[futures_table[commodity].isnull()],
    futures_filled[commodity][futures_table[commodity].isnull()],
    color="red",
    label="Filled Points",
    zorder=5
)
plt.title(f"{commodity} Weekly Futures Prices - Missing Values Handled")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

# ------------------------------------------------------------------
# ARIMA Imputation for Henry Hub Futures
# ------------------------------------------------------------------

# Fit ARIMA model using observed Henry Hub futures prices
henry_hub_obs = futures_table["HenryH_F"].dropna()
henry_hub_obs.index = pd.to_datetime(henry_hub_obs.index)
y = henry_hub_obs
model = ARIMA (y, order=(1,1,1))
fit = model.fit()

# Create full index series and insert observed values
full_index = futures_filled.index
henryh_filled = pd.Series(index=full_index, dtype=float)
henryh_filled.loc[henry_hub_obs.index] = henry_hub_obs

# Predict missing values sequentially
missing_idx = henryh_filled[henryh_filled.isna()].index
if len(missing_idx) > 0:
    pred_values = fit.predict(start=0, end=len(y)-1 + len(missing_idx))
    henryh_filled.loc[missing_idx] = pred_values[-len(missing_idx):].values


# Brent spot and Brent futures are used for visual validation as representative benchmark series; 
# identical imputation procedures were applied to all retained commodities.
```

# 6.7 Rolling Volatility

This section analyses the time-varying volatility of weekly returns across spot and futures energy markets. Rolling volatility is measured as the rolling standard deviation of weekly percentage returns over a 12-week window, providing a dynamic view of how market risk evolves through time.

Using weekly returns rather then price levels ensures that volatility is standardised and directly comparable across commodities with different price scales and market structures. The selected 12-week window approximates a quarterly horizon, balancing sensitivity to short term shocks with stability in the volatility estimates.
]


Rolling volatility patterns reveal pronounced volatility clustering, with sustained periods of elevated risk coinciding with major market disruptions. Differences in volatility magnitude and persistance across commodities reflect varying supply-demand elasticities, storage constraints, and exposure to geopolitical and macroeconomic shocks.

Heatmap visualisations complement the line plots by highlighting cross-commodity volatility regimes and synchronisation during periods of market stress. These representations provide a concise overview of relative risk dynamics across spot and futures markets.

Due to significant missing observtions in Henry Hub futures prices, volatility for this market is analysed separately using ARIMA-imputed futures returns. Presenting Henry Hub in an independent plot accounts for its distinct data characteristics while preserving continuity in the volatility assessment.

Overall, the rolling volatility analysis offers valuable insight into the evolving risk environment of energy markets and provides a robust foundation for subsequent volatility modelling, correlation analysis, and risk-management applications.

**All volatility measures are expressed in percentage return terms**


```python
# 6.8 Rolling Volatility Code

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define rolling window ( 12 weeks = quarterly volatility)
window = 12

# ------------------------------------------------------------------
# Rolling Volatility (Standard Deviation)
# ------------------------------------------------------------------ 

spots_vol = spots_returns.rolling(window).std()
futures_vol = futures_returns.rolling(window).std()

henryh_returns = henryh_filled.pct_change() * 100
henryh_vol = henryh_filled.rolling(window).std()

# ------------------------------------------------------------------
# Line Plot: Spot Rolling Volatility
# ------------------------------------------------------------------
plt.figure(figsize=(14,6))
for commodity in spots_vol.columns:
    plt.plot(spots_vol[commodity], label=commodity)
plt.title(f"Rolling Volatility of Weekly Spot Returns ({window}-Week Window)")
plt.xlabel("Date")
plt.ylabel("Volatility (%)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# ------------------------------------------------------------------
# Line Plot: Futures Rolling Volatility
# ------------------------------------------------------------------ 
plt.figure(figsize=(14,6))
for commodity in futures_vol.columns:
    plt.plot(futures_vol[commodity], label=commodity)
plt.title(f"Rolling Volatility of Weekly Futures Returns ({window}-Week Window)")
plt.xlabel("Date")
plt.ylabel("Volatility (%)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()


# ------------------------------------------------------------------
# Heatmap: Spot Rolling Volatility
# ------------------------------------------------------------------

plt.figure(figsize=(14,6))
sns.heatmap(spots_vol.T, cmap="coolwarm", cbar_kws={"label": "Volatility (%)"}, xticklabels=52)
plt.title(f"Heatmap of Rolling Volatility - Spots ({window}- Week Window)")
plt.xlabel("Date")
plt.ylabel("Spots")
plt.show()

# ------------------------------------------------------------------
# Heatmap: Spot Rolling Volatility
# ------------------------------------------------------------------
plt.figure(figsize=(14,6))
sns.heatmap(futures_vol.T, cmap="coolwarm", cbar_kws={"label": "volatility"}, xticklabels=52)
plt.title(f"Heatmap of Rolling Volatility - Futures ({window}-Week Window)")
plt.xlabel("Date")
plt.ylabel("Futures")
plt.show()
```

# 6.13 Seasonality Patterns

Seasonal behaviour is a well-documented characteristic of commodity markets, arising from recurring demand, supply, and production cycles that typically follow an annual rhythm. In energy markets, these effects are driven by factors such as seasonal travel demand, weather-related consumption, and refinery maintenance schedules.

Key seasonal drivers include:

1. Increased summer travel demand affecting jet fuel and gasoline consumption
2. Higher winter heating demand influencing natural gas and Hnery Hub prices
3. Planned refinery maintenance impacting crude oil benchmarks and refined product supply 
4. Broader industrial and agricultural cycles indirectly affecting diesel and energy demand

To identify seasonal effects, weekly price series are first converted into monthly averages. This transformation smooth short-term price fluctuations and highlights underlying seasonal patterns. Seasonality is then analysed by grouping observations by calender month and computing the average price for each month across the sample period.

Three seasonal price curves are produced:

1. **Spot Prices** - Illustrating how physical market conditions evolve throughout the year
2. **Futures Prices** - Capturing seasonal patterns embedded in market expectations of future supply and                           demand
3. **Henry Hub Futures (ARIMA-imputed)** - presented separately due to higher volatility and a distinct                                               price scale relative to other futures contracts

Seasonal curves allow identification of systematic price peaks and troughs across calender months. Recognising these recurring patterns is important for improving forecasting accuracy, informing timing decisions and supporting trading and hedging strategies.

Given the substantial differences in price levels across energy commodities, dual y-axes are eployed in the seasonal charts. This ensures that lower-priced commodities remain visually interpretable and that seasonal dynamics can be meaningfully compared across markets.


```python
# 6.14 Seasonality Patterns Code 

import pandas as pd
import matplotlib.pyplot as plt

# Ensure datetime index 
spots_filled.index = pd.to_datetime(spots_filled.index)
futures_filled.index = pd.to_datetime(futures_filled.index)
henryh_filled.index = pd.to_datetime(henryh_filled.index)

# Compute monthly averages for seasonality 
spots_monthly = spots_filled.groupby(spots_filled.index.month).mean()
futures_monthly = futures_filled.groupby(futures_filled.index.month).mean()
henryh_monthly = henryh_filled.groupby(henryh_filled.index.month).mean()

# Reset index to 1-12 for months 
spots_monthly.index = range(1,13)
futures_monthly.index = range(1,13)
henryh_monthly.index = range(1,13)

# ------------------------------------------------------------------
# Seasonality Plot: Spot Prices (Dual Y-Axis) 
# ------------------------------------------------------------------

spot_high = ["Brent_S", "WTI_S"]
spot_low = [col for col in spots_monthly.columns if col not in spot_high]
plt.figure(figsize=(14,6))
ax1 = plt.gca()

# High-price commodities (left axis) 
for col in spot_high:
    if col in spots_monthly.columns:
        ax1.plot(spots_monthly.index, spots_monthly[col], label=col, linewidth=2)

ax1.set_xlabel("Month")
ax1.set_ylabel("High-price commodities")
ax1.set_xticks(range(1,13))
ax1.grid(True, alpha=0.3)

# Low-price commodities (right axis) 
ax2 = ax1.twinx()
for col in spot_low:
    ax2.plot(spots_monthly.index, spots_monthly[col], linestyle="--", label=col)

ax2.set_ylabel("Low-price commodities")

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
plt.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

plt.title("Seasonality Pattern - Monthly Average Spot Prices")
plt.show()

# ------------------------------------------------------------------
# Seasonality Plot: Futures Prices (Dual Y-Axis + Henry Hub)
# ------------------------------------------------------------------ 

fut_high = ["Brent_F"]
fut_low = [col for col in futures_monthly.columns if col not in fut_high]
plt.figure(figsize=(14,6))
ax1 = plt.gca()

# High-price futures (left axis)
for col in fut_high:
    if col in futures_monthly.columns:
        ax1.plot(futures_monthly.index, futures_monthly[col], linewidth=2, label=col)

ax1.set_xlabel("Month")
ax1.set_ylabel("High-price futures")
ax1.set_xticks(range(1,13))
ax1.grid(True, alpha=0.3)

# Low-price futures + Henry Hub (right axis)
ax2 = ax1.twinx()
for col in fut_low:
    ax2.plot(futures_monthly.index, futures_monthly[col], linestyle="--", label=col)

# Henry Hub ARIMA
ax2.plot(
    henryh_monthly.index,
    henryh_monthly.values,
    color="red",
    linewidth=2,
label="HenryH_F (ARIMA)"
)

ax2.set_ylabel("Low-price futures (incl. Henry Hub)")

# Combine legends
l1, lab1 = ax1.get_legend_handles_labels()
l2, lab2 = ax2.get_legend_handles_labels()
plt.legend(l1 + l2, lab1 + lab2, loc="upper left")

plt.title("Seasonality Pattern - Monthly Average Futures Prices ")
plt.show()
```

**Market Structure Interpretations:**

**Backwardation**(Front-month > Next-month): 
Signals a tight market. discourages storage, and incentivizes immediate physical delivery
**Contango** (Front-month < Next-month): 
Indicates surplus supply, incentivizes storage, and reflects positive carry costs.

**Data Limitation:**
In this analysis, **only continuous front-month futures series** are available for the relevant commodities. Since prices for multiple simultaneous delivery maturities are not observed, **true calender spreads cannot be computed** without introducing methodological distortions.

Constructing calendar spreads from a single rolling front-month contract would conflate:
- Contract roll effects
- Term structure dynamics
- Spot–futures convergence
and would therefore not provide a reliable measure of market structure.

**Importance of Calendar Spreads**
Despite this limitation, calendar spreads remain a core analytical tool in energy markets, including:
- Crude oil (Brent, WTI)

# 9.1 Spread Volatility and Extreme Price Movements 
Energy markets are defined not only by price levels, but by the **volatility and extremity of price movements**, particularly across key spreads that underpin trading, refining margins, and risk management decisions. For integrated energy companies, understanding how spreads behave under different market conditions is critical for **portfolio optimisation, hedging strategies, and operational resilience**.

This section analyzes volatility dynamics and extreme price movements across three strategically important spreads:

- **Brent-WTI Spread**: Capturing regional crude oil dislocations and logistical constraints
- **Gasoline crack spread**: Reflecting downstream refining economics and demand sensitivity
- **3-2-1 crack spread**: Serves as a proxy for overall refinery margin performance

Rather than analyzing spot prices in isolation, this approach emphasizes **relative value** and risk, aligning with how real-world trading and commercial teams evaluate market exposure.

To quantify market risk over time, a **26-week rolling window** is applied. This period is long enough to smooth short-term noise while remaining responsive to structural shifts driven by macroeconomic events, supply disruptions, or demand shocks. Within each window, the analysis captures:

- **Rolling volatility (Standard Deviation)**: Measures market uncertainty
- **Maximum and minimum Spread Movements**: Highlighs periods of extreme stress or opportunity

By combining rolling volatility with extreme value analysis, this framework enables a deeper assessment of **when spreads become unstable**, how persistent that instability is, and which segments of the energy value chain are most exposed. The results provide actionable insight into periods where **risk controls, hedging intensity, or capital allocation decisions** would be most critical.

The following section implements this methodology programmatically and visualises how volatility and extreme movements evolve across time for each spread.

```python
# 9.2 Spread Volatility and Extreme Price Movements Code

import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------------------------------------------
# Align date indexes across datasets
# ------------------------------------------------------------------  
common_index = spots_filled.index.intersection(ref_prod_spreads.index)
spots_filled = spots_filled.reindex(common_index)
ref_prod_spreads = ref_prod_spreads.reindex(common_index)


# ------------------------------------------------------------------
# Construct a unified DataFrame of key energy spreads
# ------------------------------------------------------------------ 
vol_df = pd.DataFrame({
    "Brent_WTI_Spread": spots_filled["Brent_S"] - spots_filled["WTI_S"],
    "Gasoline_Crack": ref_prod_spreads["Gasoline_Crack"],
    "Crack_3_2_1": ref_prod_spreads["Crack_3_2_1"]
})

# ------------------------------------------------------------------
# Define rolling window ( 26 weeks = 6 months)
# ------------------------------------------------------------------  
rolling_window = 26 #around 6 months
vol_metrics = pd.DataFrame(index=vol_df.index)

# Compute rolling volatility and extreme movements for each spread
for spread in vol_df.columns:
    vol_metrics[f"{spread}_RollingStd"] = vol_df[spread].rolling(rolling_window).std()
    vol_metrics[f"{spread}_MaxMove"] = vol_df[spread].rolling(rolling_window).max()
    vol_metrics[f"{spread}_MinMove"] = vol_df[spread].rolling(rolling_window).min()

# ------------------------------------------------------------------
# Plotting Data
# ------------------------------------------------------------------ 

# Plot rolling volatility for Brent-WTI spread
plt.figure(figsize=(12,5))
plt.plot(vol_metrics.index, vol_metrics["Brent_WTI_Spread_RollingStd"], label="Brent-WTI Volatility", color="Blue")
plt.title("Rolling 26-Week Volatility of Brent-WTI Spread")
plt.xlabel("Date")
plt.ylabel("$/bbl")
plt.legend()
plt.grid(True)
plt.show()

# Plot rolling volatility for refined product cracks
plt.figure(figsize=(12,5))
plt.plot(vol_metrics.index, vol_metrics["Gasoline_Crack_RollingStd"], label="Gasoline Crack Volatility", color="green")
plt.plot(vol_metrics.index, vol_metrics["Crack_3_2_1_RollingStd"], label="3-2-1 Crack Volatility", color="orange")
plt.title("Rolling Volatility of Gasoline & 3-2-1 Crack Spreads")
plt.xlabel("Date")
plt.ylabel("$/bbl")
plt.legend()
plt.grid(True)
plt.show()


# ------------------------------------------------------------------
# Summary Statistics
# ------------------------------------------------------------------ 
display(vol_metrics[[f"{s}_MaxMove" for s in vol_df.columns] + [f"{s}_MinMove" for s in vol_df.columns]].describe())  
 
```

# 9.4 Rolling Drawdowns and Peak-to-Trough Risk
While volatility captures the magnitude of spread movements, it does not fully reflect **how sustained losses or adverse deviations accumulate over time**. Rolling drawdowns provide a measure of **peak-to-trough risk**, highlighting periods where spreads experience prolonged declines and evaluating the persistence of downside exposure.

This analysis focuses on two key spreads:
- **Brent-WTI Spread:** Captures regional crude price differentials and informs arbitrage and sourcing decisions.
- **Gasoline Crack Spread:** Reflects refining margins and the sensitivity of downstream profitability to changes in crude and product prices.

The methodology uses a **26-week rolling window** to calculate cumulative changes in spreads. Within each window:
1. The **rolling cumulative change** is computed.
2. The **rolling peak** is identified.
3. The **drawdown** is calculated as the difference between the current value and the peak, representing the **worst-case cumulative decline over the window**.

The resulting plots illustrate the **magnitude and duration of drawdowns**, providing insight into periods of heightened risk for both crude and refining spreads. Summary statistics quantify the typical and extreme drawdowns, helping to inform risk management, **hedging strategies, and operational decisions** in the context of integrated energy trading and portfolio management.


```python
# 9.5 Rolling Drawdowns and Peak-to-Trough Risk Code

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

drawdown_df = pd.DataFrame({
    "Brent_WTI_Spread": spots_filled["Brent_S"] - spots_filled["WTI_S"],
    "Gasoline_Crack": ref_prod_spreads["Gasoline_Crack"]
})

spread_returns = drawdown_df.diff()
rolling_window = 26
rolling_cum = spread_returns.rolling(rolling_window).sum()

rolling_peak = rolling_cum.cummax()
rolling_drawdown = rolling_cum - rolling_peak

plt.figure(figsize=(12,5))
plt.plot(rolling_drawdown.index, rolling_drawdown["Brent_WTI_Spread"], label="Brent-WTI Drawdown", color="blue")
plt.title("Rolling 26-week Drawdown: Brent-WTI Spread")
plt.xlabel("Date")
plt.ylabel("Drawdown $/barrel")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(12,5))
plt.plot(rolling_drawdown.index, rolling_drawdown["Gasoline_Crack"], label="Gasoline Crack Drawdown", color="green")
plt.title("Rolling 26-week drawdown: Gasoline Crack Spread")
plt.xlabel("Date")
plt.ylabel("Drawdown $/barrel")
plt.grid(True)
plt.legend()
plt.show()

drawdown_summary = rolling_drawdown.describe()
display(drawdown_summary)
```

# 9.7 Risk Adjusted Performance 

This section evaluates the **risk-adjusted performance** of key energy spreads, providing a standardized metric to compare their returns relative to their volatility. It allows traders and analysts to assess which spreads offer the most efficient trade-off between **return and risk**.

## Spreads Analyzed

-**Brent-WTI Spread**: captures regional crude price differentials 
-**Gasoline Crack Spread**: reflects refining margin on gasoline 
-**3-2-1 Crack Spread**: proxy for intergrated refinery margin combining gasoline and diesel

## Methology

1. **Spread Return Calculation**
   - Weekly spread chnages are used as the return measure in $/bbl
   - Absolute price chnages are appropiate since spreads are already denominated in monetary and reflect direct **P&L impact**

2. **Annualization**
   - Weekly mean returns are annualized by multiplying by 52 weeks
   - Weekly volatility (standard deviation) is annualized by multiplying by square-root 52
   - Annualization allows **consistent comparison between spreads** and with other instruments
  
3. **Sharpe ratio computation**
   - The sharpe ratio is calculated as the **annualized return / annualized volatility**
   - Assuming **risk-free rate = 0**, consistent with relative-valuespread trading
   - No transaction costs or leverage are applied, since the goal is to assess **intrinsic spread
     behaviour** , not simulate a trading strategy

4. **Interpretation**
   - A higher sharpe ratio indicates a **more favorable risk-adjusted performance**, meaning the spread
     provides greater return per unit of risk
   - comparing upstream (Brent-WTI) and downstream (Gasoline and 3-2-1 crack) spreads highlights which
     part of the energy value chain offers **more stable risk-adjusted opportunities**

**Purpose of this Section:**
- Provide a **quantative, risk-adjusted benchmark** for spreads 
- Identify spreads that deliver **consistent returns relative to volatility** 
- Inform **trading, hedging and capital allocation decisions** without making speculative assumptions 
- Demonstrates understanding of **risk-return trade-offs in intergrated energy markets** 


```python
# 9.8 Risk Adjusted Performance Code

import numpy as np
import pandas as pd

# ------------------------------------------------------------------
# Construct DataFrame of key spreads
# ------------------------------------------------------------------
sharpe_df = pd.DataFrame({
    "Brent_WTI_Spread": spots_filled["Brent_S"] - spots_filled["WTI_S"],
    "Gasoline_Crack": ref_prod_spreads["Gasoline_Crack"],
    "Crack_3_2_1": ref_prod_spreads["Crack_3_2_1"]
}).dropna()

# ------------------------------------------------------------------
# Calculate weekly spread changes as returns in $/bbl
# ------------------------------------------------------------------ 
week_return = sharpe_df.diff().dropna()

# ------------------------------------------------------------------
# Annualize return and volatility
# ------------------------------------------------------------------
annualized_return = week_return.mean() * 52
annualized_volatility = week_return.std() * np.sqrt(52)

# ------------------------------------------------------------------
# Compute Sharpe ratio (risk-free rate = 0)
# ------------------------------------------------------------------
sharpe_ratio = annualized_return / annualized_volatility

# ------------------------------------------------------------------
# Summary Table
# ------------------------------------------------------------------
sharpe_summary = pd.DataFrame({
    "Annualized Return ($/bbl)": annualized_return,
    "Annualized Volatility ($/bbl)": annualized_volatility,
    "Sharpe Ratio": sharpe_ratio
})

# Display results
display(sharpe_summary)
```
